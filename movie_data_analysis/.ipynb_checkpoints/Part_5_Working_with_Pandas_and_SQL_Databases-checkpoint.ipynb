```python
import sqlite3 as sq3
```

```python
con = sq3.connect("movies.db")
# with connect we can connect to an existing databse in our directory
# and if the database name passed does not exixt it creates 
# a new database with this name 
```

```python
con
#con is the connection to the databse movies.db 
```

```python
con.execute("select * from sqlite_master").fetchall()
# to execute a query 
```

```python
#always close connection 
con.close()

# to execute a query the connection should be open 
```

```python
import pandas as pd
import json
import sqlite3 as sq3
```

```python
# opening thejson file and storing data in the variable data

with open("some_movies.json") as f:
    data = json.load(f)
```

```python
data
```

```python
df = pd.json_normalize(data, sep="_")
```

```python
df.head(5)
```

```python
# creating a data frame with few attributes of df 
movies = df[["id","title","revenue","budget","belongs_to_collection_name","release_date"]].copy()
```

```python
movies
```

```python
movies.info()
```

```python
movies.release_date = pd.to_datetime(df.release_date)
```

```python
movies.revenue = movies.revenue/1000000
```
