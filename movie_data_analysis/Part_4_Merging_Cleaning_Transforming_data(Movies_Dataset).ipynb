```python
#now we have to clean and transform cast column 
```

```python
import ast 
import numpy as np
```

```python
df.cast[0]
```

```python
# now above is a stringified json data or list 
#so we extract the lists 
```

```python
df.cast = df.cast.apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)
```

```python
df.cast[1]
```

```python
# now each row of cast has multplie dictionaries containg info,
# we can convert each row into a data frame 

pd.DataFrame(df.cast[0])
# for each dictionary of the cast row we get a row in this new dataframe
# for each row of cast one dataframe 
```

```python
# now we do not need all the info of the cast we just need 
# the cast names column where we will have all names of cast 
#seperated by pipe 

# also we can add a column in our df table specifying the cast size 
# now cast size for a movie would be the number of dictionaries for each row 

df["cast_size"] = df.cast.apply(lambda x: len(x))
```

```python
df.cast_size.value_counts(dropna = False).head(50)
```

```python
df.cast = df.cast.apply(lambda x: "|".join(i['name'] for i in x) if isinstance(x, list) else np.nan)
```

```python
df.cast
```

```python
df.cast.replace(" ", np.nan , inplace = True)
```

```python
df.cast.value_counts()
```

```python
# cleaning and transforming crew column \

df.crew[0]
```

```python
df.crew = df.crew.apply(lambda x : ast.literal_eval(x) if isinstance(x,str) else np.nan)
```

```python
df.crew[0]
```

```python
pd.DataFrame(df.crew[0])
```

```python
df["crew_size"] = df.crew.apply(lambda x: len(x))
```

```python
df.crew_size.value_counts(dropna = False).head(50)
```

```python
# we only need the director 
# each crew row is a list of dictionaries with each dict for each crew 
# member of that movie 
# so we iterate the list of each movies crew and extract the directors name 
# from there 

```

```python
# user defined function 

def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
        return np.nan
```

```python
#apply on each crew row to parse all the dict of the list 

df["director"] = df.crew.apply(get_director)
```

```python
df.director.value_counts(dropna=False).head(50)
```
