```python
df_best.Budget.fillna(0, inplace = True)
df_best.Votes.fillna(0, inplace = True)
```

```python
# function 
```

```python
def best_worst(n, by, ascending= False, min_bud =0 , min_votes = 0):
    
    df2 = df_best.loc[(df_best.Budget >= min_bud) & \
                      (df_best.Votes >= min_votes), \
                      ["",by]]\
        .sort_values(by = by, ascending = ascending).head(n).copy()
    
    
    return HTML(df2.to_html(escape=False))
```

```python
# Highest Revenue
best_worst(5, by= "Revenue")
```

```python
#Highest Budget
best_worst(5, by="Budget")
```

```python
#Highest Profit
best_worst(5,"Profit")
```

```python
# top 5 lowest rating
best_worst(5, "Average Rating", ascending = True, min_votes = 100)
```

```python
best_worst(5, "Average Rating", ascending = True, min_votes = 20, min_bud = 20)
```
